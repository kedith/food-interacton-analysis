```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
category_df = pd.read_csv('../initial_data/categories.csv', names=['drug_id', 'category'])
category_df = category_df[1:] #remove first line containing the header
```

```python
category_df.head()
```

Number of unique category values 

```python
category_df['category'].nunique()
```

Number of unique drug ids

```python
category_df['drug_id'].nunique()
```

```python
category_df.shape
```

```python
category_df['category'].value_counts()[:20]
```

```python
plt.figure(num=None, figsize=(20, 8))
plt.rc('xtick',labelsize=15)
plt.rc('ytick',labelsize=15)
plt.title('Most frequent categories', fontsize=20)
category_df['category'].value_counts()[:20].plot(kind='barh')
#plt.savefig('Most frequent categories.png',dpi=300, bbox_inches = "tight")
plt.show()
```

```python
food_df = pd.read_csv('../initial_data/food_interactions.csv')
food_df.rename(columns=food_df.iloc[0]).drop(food_df.index[0]);
```

```python
food_df.head()
```

```python
food_df['description'] = food_df['description'].str.replace(r'.$', '')
```

```python
food_df.head()
```

```python
plt.figure(num=None, figsize=(10, 6))
plt.title('Most frequent food interactions', fontsize=15)
food_df['description'].value_counts()[:20].plot(kind='barh')
plt.savefig('Most frequent descriptions.png',dpi=300, bbox_inches = "tight")
plt.show()
```

```python
extended_food_df=pd.merge(category_df, food_df, on='drug_id', how='inner')
```
